## Storytelling Data Visualization on Exchange Rates
### Guided Project - Dataquest

The goal of this project is analyse the euro (€) daily exchange rates between 1999 and 2021.

The exchange rate of the euro to the US dollar is 1.5 (1 € = 1.5 $).

Daria Chemkaeva put together the data set and made it available on [kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020). The data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on October 2022.

Let's start what kind of data we can find in this dataset:

```python
#Import pandas
import pandas as pd

#Import dataset
exchange_rates = pd.read_csv('euro-daily-hist_1999_2022.csv',sep=',')

exchange_rates.head()
```

```python
exchange_rates.info()
```

There are 6147 rows representing a day and 41 columns (one for the date and other 40 for each country currency).

There are some countries with some null values as Bulgary, Brazil, China, Cyprus, Estonia, Greece, Croatia, Israel, India, Iceland, Lithuania, Latvia, Malta, Romania, Russia, Slovenia, Slovakia and Turkey.

Most columns are object type instead of float (just Iceland, Turkey and Romania). Even Period\Unit is not a time type.

### Cleaning data

Our study will be focused on **euros - US dollar** exchanges so we are going to make the following changes to the dataset:
- To rename the [US dollar ] and Period\Unit: columns to something easier to type — US_dollar and Time.
- To change the Time column to a datetime data type.
- To sort the values by Time in ascending order.
- To reset the index (and drop the initial index).

```python
#Renaming the [US dollar ] and Period\Unit: columns to something easier to type — US_dollar and Time.
exchange_rates.rename({"[US dollar ]":"US_dollar"},axis=1,inplace=True)
exchange_rates.rename({"Period\\Unit:":"Time"},axis=1,inplace=True)

#Chaging Time column to datetime type
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

exchange_rates.info()
```

```python
#Sorting the values by Time in ascending order.
exchange_rates = exchange_rates.sort_values('Time',ascending=True)

exchange_rates.head()
```

```python
#Resetting the index (and drop the initial index)
exchange_rates.reset_index(drop=True, inplace=True)

exchange_rates.head()
```

We are going to isolate `Time` and `US_dollar` columns:

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']]
euro_to_dollar.head()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

There are 3675 different values of exchange rates. But there is a value that is not a number ('-') which is making our column Object type. We are going to drop those 62 rows and convert this column in float type.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

### Visualizing Data

```python
#import matplotlib and enable Jupyter to display graphs
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline
```

#### Daily variation in the exchange rate

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.xlabel('Date')
plt.ylabel('Exchange rate')
plt.title('Daily variation in the € - $ exchange rate', weight='bold')
plt.show()
```

We can see this daily variation with a rolling mean (also known as the moving average).

*The value is the mean between the value of that day and the value of the previous day. For the third day, the value is the mean between the third and the second day. For the seventh day, the value is the mean between the seventh and the sixth day, and so on.*

*We calculated the rolling mean using values from two days — the current day and the previous day. We call the number of days used in the calculation the rolling window (or moving window). If the rolling window is three, then we calculate the mean between the value of the current day and the values of the previous two days.*

There is a pandas method to calculate this: pandas.Series.rolling(days).mean()

```python
fig,((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(nrows=3, ncols=2, figsize=(12,10))
rolling_windows = [0,7,30,50,100,365]
axes = [ax1,ax2,ax3,ax4,ax5,ax6]
for ax,w in zip(axes,rolling_windows):
    if w != 0:
        ax.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(w).mean())
        ax.text(16800,euro_to_dollar['US_dollar'].rolling(w).mean().max()-0.05,'Rolling window: '+str(w), size=8, weight='bold')
    else:
        ax.plot(euro_to_dollar['Time'],
        euro_to_dollar['US_dollar'])
        ax.text(16800,euro_to_dollar['US_dollar'].max()-0.05,'Rolling window: '+str(0), size=8, weight='bold')
    
ax1.text(10000,1.7,'Daily variation in the € - $ exchange rate', weight = 'bold', size=14)
plt.show()
```

```python
#Rolling means for the US_dollar column using a moving window of 30 days.
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

### Storytelling

Some ideas:
- We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
- We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 and 2009 for comparison. We can use a line plot.
- We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

Let's check how the euro-dollar changed during the **2007-2008 financial crisis** using:
- Data-ink ratio maximum.
- Visual patterns using Gestalt principles.
- Pre-attentive attributes to guide the audience's attention.

#### Financial Crisis Example

```python
#Isolating data
period_total = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <= 2009)]
period_crisis = euro_to_dollar[(euro_to_dollar.Time.dt.year >= 2007) & (euro_to_dollar.Time.dt.year <= 2008)]

#style
style.use('fivethirtyeight')

#plot
fig,ax = plt.subplots(figsize=(8,4))
ax.plot(period_total['Time'],period_total['rolling_mean'], color='#A6D785', linewidth=1)
ax.plot(period_crisis['Time'],period_crisis['rolling_mean'], color='#e23d28',linewidth=3)

#removing ticks labels

ax.set_xticks(['2006-01-01','2007-01-01','2008-01-01','2009-01-01','2009-12-31'])
ax.set_xticklabels(['2006','2007','2008','2009','2010'], color='grey')

ax.set_yticks([1.2,1.3,1.4,1.5])
ax.set_yticklabels(['1.2','1.3','1.4','1.5'], color='grey')

#title and subtitle

ax.text(13000,1.65,'Euro-USD rate peaked at 1.59 during 2007-2008\'s financial crisis', weight='bold')
ax.text(13000,1.62,'Euro-USD exchange rates between 2006 and 2010', size=12)

#highlight max - enclosure
max_rate = period_crisis[period_crisis['rolling_mean'] == period_crisis['rolling_mean'].max()]
ax.axvline (x=14035,c='grey',alpha=0.2, linewidth=70, ymin=0.08)

#signature

ax.text(13000,1.1,'©DATAQUEST'+100*' '+'Source: European Central Bank',color='white',backgroundcolor='grey', size=10)

plt.show()
```
